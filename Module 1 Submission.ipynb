## Final analysis code

```python
#conda install -c conda-forge termcolor

from matplotlib import pyplot as plt
import statistics
import pandas as pd
import csv
import warnings
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

try:
    from termcolor import colored
except Exception:
    def colored(s, *_args, **_kwargs): return s

class Patient:
    all_patients = []
    education_lvl = {}

    def __init__(self, DonorID, ABeta40: float , ABeta42: float, tTau: float, pTau: float, pH: float):
        self.DonorID = DonorID
        self.ABeta40 = ABeta40
        self.ABeta42 = ABeta42
        self.tTau = tTau
        self.pTau = pTau
        self.pH = pH
        self.sex = None
        self.death_age = None
        self.ed_lvl = None
        self.cog_stat = None
        self.age_symp_on = None
        self.age_diag = None
        self.head_inj = None
        self.thal_score = None
        Patient.all_patients.append(self)

    def __repr__(self):
        return (f"{self.DonorID} | sex: {self.sex} | ABeta40 {self.ABeta40} | "
                f"tTau {self.tTau} | pTau {self.pTau} | Death Age {self.death_age} | "
                f"Thal Score {self.thal_score} | Brain pH {self.pH}")

    def get_id(self): return self.DonorID
    def get_ABeta42(self): return self.ABeta42
    def get_thal(self): return self.thal_score
    def get_death_age(self): return self.death_age
    def get_pH(self): return self.pH

    @staticmethod  # now actually on the class
    def classify_pH(pH: float):
        if pH < 7:
            return "Acidic"
        elif pH == 7:
            return "Neutral"
        else:
            return "Basic"

    @classmethod
    def combine_data(cls, filename: str):
        # join by Donor ID, not row index
        id_to_patient = {p.DonorID: p for p in cls.all_patients}
        with open(filename, encoding="utf8") as f:
            reader = csv.DictReader(f)
            for row in reader:
                pid = row.get("Donor ID")
                if not pid or pid not in id_to_patient:
                    warnings.warn(f"ID {pid} missing in initial data.")
                    continue
                p = id_to_patient[pid]

                if row.get("Sex"): p.sex = row["Sex"]
                if row.get("Age at Death"): p.death_age = int(row["Age at Death"])
                if row.get("Highest level of education"): p.ed_lvl = row["Highest level of education"]
                if row.get("Cognitive Status"): p.cog_stat = row["Cognitive Status"]
                if row.get("Age of onset cognitive symptoms"): p.age_symp_on = int(row["Age of onset cognitive symptoms"])
                if row.get("Age of Dementia diagnosis"): p.age_diag = int(row["Age of Dementia diagnosis"])
                if row.get("Known head injury"): p.head_inj = row["Known head injury"]
                if row.get("Thal"):
                    # handle things like "Thal 3"
                    parts = row["Thal"].split()
                    if parts and parts[-1].isdigit():
                        p.thal_score = int(parts[-1])
                if row.get("Brain pH"):
                    try:
                        p.pH = float(row["Brain pH"])  # keep as float
                    except ValueError:
                        pass  # leave existing pH

    @classmethod
    def instantiate_from_csv(cls, filename: str, other_file: str):
        with open(filename, encoding="utf8") as f:
            reader = csv.DictReader(f)
            for row in reader:
                Patient(
                    DonorID = row['Donor ID'],
                    ABeta40 = float(row['ABeta40 pg/ug']),
                    ABeta42 = float(row['ABeta42 pg/ug']),
                    tTau = float(row['tTAU pg/ug']),
                    pTau = float(row['pTAU pg/ug']),
                    pH = float(row['Brain pH']) if row.get('Brain pH') else np.nan
                )
        cls.all_patients.sort(key=Patient.get_id)
        cls.combine_data(other_file)

    @classmethod  # now actually on the class
    def group_by_pH(cls):
        cls.pH_groups = {"Acidic": [], "Neutral": [], "Basic": []}
        for patient in cls.all_patients:
            if patient.get_pH() is None or (isinstance(patient.get_pH(), float) and np.isnan(patient.get_pH())):
                continue
            category = cls.classify_pH(patient.get_pH())
            cls.pH_groups[category].append(patient)

    @classmethod  # now actually on the class
    def subsort_abeta42(cls):
        for key in cls.pH_groups:
            cls.pH_groups[key].sort(key=lambda p: p.get_ABeta42())

# ----- build patients -----
Patient.instantiate_from_csv('UpdatedLuminex.csv', 'UpdatedMetaData.csv')

# Optional sanity check
# for p in Patient.all_patients: print(p)

# ----- group & sort -----
Patient.group_by_pH()
Patient.subsort_abeta42()

for key in Patient.pH_groups:
    print(colored(key, "red"))
    for patient in Patient.pH_groups[key]:
        print(patient)
    print()

# ----- compute stats & plot -----
# Extract ABeta42 values for each pH group
abeta_groups = [ [p.get_ABeta42() for p in patients] for patients in Patient.pH_groups.values() ]

# Run one-way ANOVA

f_stat, p_value = stats.f_oneway(*abeta_groups)

print("F statistic: ", f_stat)
print("P value: ", p_value)

avg_abeta42_by_pH = {}
std_abeta42_by_pH = {}

#compute stats and plot
for group, patients in Patient.pH_groups.items():
    abeta_values = [p.get_ABeta42() for p in patients]
    if abeta_values:
        avg_abeta42_by_pH[group] = np.mean(abeta_values)
        std_abeta42_by_pH[group] = np.std(abeta_values, ddof=1)  # sample stddev
    else:
        avg_abeta42_by_pH[group] = 0.0
        std_abeta42_by_pH[group] = 0.0

# stable order
groups = ["Acidic", "Neutral", "Basic"]
values = [avg_abeta42_by_pH.get(g, 0.0) for g in groups]
errors = [std_abeta42_by_pH.get(g, 0.0) for g in groups]

plt.figure(figsize=(8, 6))
plt.bar(groups, values, yerr=errors, capsize=8, edgecolor='black')
plt.xlabel('pH Group')
plt.ylabel('Average A42 Level (pg/g)')
plt.title('Average A42 Levels by Brain pH Group (1 SD)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.text(2, 220, f"One way ANOVA: p= {p_value:.3f}", ha='right', va='top', fontsize=12)
plt.savefig('abeta42_by_pH.png')  
plt.show()
print("Saved figure to abeta42_by_pH.png")

# 9) Get patient attributes for scatter plot

Brain_pH_list = []
ABeta42_list = []

for patient in Patient.all_patients:
    # Only include patients with valid (non-NaN) values
    if patient.pH is not None and not np.isnan(patient.pH) and patient.ABeta42 is not None and not np.isnan(patient.ABeta42):
        Brain_pH_list.append(patient.pH)
        ABeta42_list.append(patient.ABeta42)

# 10) Visualize on scatter plot

plt.figure(figsize=(7, 5))
plt.scatter(Brain_pH_list, ABeta42_list, color='blue')
plt.xlabel('Brain pH')
plt.ylabel('ABeta42')
plt.title('Scatter Plot of Brain pH vs ABeta42')
plt.grid(True)
plt.tight_layout()
plt.show()


#11 export data to a .csv file
import pandas as pd
print(Brain_pH_list)
print(ABeta42_list)

#create a DataFrame
df = pd.DataFrame({
    'Brain pH': Brain_pH_list,
    'ABeta42': ABeta42_list})
#Write to CSV
df.to_csv('patient_data.csv', index = False)
print("CSV file 'patient_data.csv' has been created.")

#12 load libraries for a linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#13 load data set for linear regression
df = pd.read_csv("patient_data.csv")

#14 update these varibale names to match exactly your.csv file headers

x = df["Brain pH"].values.reshape(-1,1)
y = df['ABeta42'].values

#15 perform the linear regression

model = LinearRegression()
model.fit(x,y)

slope = model.coef_[0]
intercept = model.intercept_
r2 = model.score(x,y)

#16 make a scatterplot with the equation for the linear regression and the R-squared value
plt.scatter(x,y, color = "blue")
plt.plot(x, model.predict(x), color = "red")

#annotate equation
equation = f"y = {slope:.2f}x + {intercept:.2f}\nR^2={r2:.2f}"
plt.text(x.min(), y.max(), equation, color="red", fontsize=12, verticalalignment = 'top')

#annotate scatterplot with labels and title
plt.xlabel("Brain pH")
plt.ylabel("ABeta42")
plt.title("Brain pH vs ABeta42")
plt.show()
```
